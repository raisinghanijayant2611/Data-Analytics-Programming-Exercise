### 3 b)

```python
def words(text):
    text = re.sub('\.', ' ', text) #to handle "i...i chose the road least followed "
    text = re.sub('[^\w ]', '', text)
    text = text.split()
    return text

words('"The c_lever fox ~ ~ jumped 3over the 1lazy...dog"')
```

### 3 c)

```python

def postingslist(quote):   #for a given quote, creates dictionary of words in quote 
                              #and respective word count
    dic = {}
    word = words(quote)
    distinct = list(set(word))   #distinct words in quote
    for i in range(len(distinct)):   #iterate over each distinct word in quote
        wordcount = 0                  
        for j in range(len(word)):   #iterate over all words in quote
            if(distinct[i]==word[j]):   #if the word in quote matches the distinct word,
                  wordcount+=1             #add one to word count
        dic.update({distinct[i]:wordcount})   #add dictionary element for the distinct word
    return dic

pldic = {}
for i in range(len(allquotes)):
    pldic.update({allquotes[i]: postingslist(allquotes[i])})   #for each quote, create dictionary
print pldic
```

### 3 d)

```python
def revpostingslist():   #for a given distinct word in text file, create dictionary of quotes that
                            #contain that word and the word count in that quote
    dic = {}
    master=[]   #used to get distinct words in allquotes
    for i in range(len(allquotes)):   #these for loops parse all quotes and add each word to list
        hold = words(allquotes[i])
        for j in range(len(hold)):
            master.append(hold[j])
    
    distinct = list(set(master)) #keeps only distinct words in a master list
    
    for word in distinct:   #iterate over each distinct word (builds dictionary key)
        holdd={}
        for quote, l in pldic.iteritems():   #this loop creates each value of the reverse dictionary
            for w, count in l.iteritems():   
                if word==w:                  #if the word in the quote matches the word key,
                    holdd.update({quote:count})   #add element to {quote: word count} dictionary
        dic.update({word:holdd})   #after iterating through each quote that has the word key,
                                      #add the {quote: word count} dictionary toreverse dictionary
            
    return dic
    
rpldic = revpostingslist()
print rpldic['i']
```

### 3 e)

```python
def tfidf(word, quote):
    w = rpldic[word][quote]   #word count
    n=0
    for i, count in pldic[quote].iteritems():   #loop searches the quote for highest freq word
        if n<count:
            n=count

    qnumt = len(pldic)   #gets total number of quotes
    qnumw = len(rpldic[word])   #gets total number of quotes that contain word
    
    tf = 1.*w/n
    idf = 1.*math.log(1.*qnumt/qnumw)
    return tf*idf

tfidf('entertainer', '"an actor is at most a poet and at least an entertainer. - marlon brando"')
```

### 3 f)

```python
def qsearch(word): 
    dic={}
    
    if rpldic.get(word, 'No Word')=='No Word':   #catch for when word does not exist in reverse dict
        return 'Invalid Word'
    
    for i in rpldic[word]:   #for a given word, look through all quotes, calculate respective tf-idf
        dic.update({i: tfidf(word, i)})
            
    return dic

qsearch('other')
```

### 3 g)

```python
def qsearchl(l):
    dic={}
    
    for i in l:
        if rpldic.get(i, 'No Word')=='No Word':  #catch for when word does not appear in reverse dict
            return 'Invalid Word'
        dich = qsearch(i)
        for j in dich:
            if dic.get(j)==None:   #if the quote does not exist in the dictionary dic yet,
                dic.update({j:dich[j]})   #add the quote to the dictionary
            else:                  #if the quote does exist in the dictionary already,
                dic[j] = dic[j]+dich[j]   #add the new calculated tf-idf to the original
    return dic

qsearchl(['the'])
```
